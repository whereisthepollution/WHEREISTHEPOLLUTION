# Parameters:
- Selected station: 24 
- MAGNITUD: 8, corresponding to $NO_2$

```python
station_id = 24
magnitud = 8
```

```python
df = df[(df['ESTACION'] == station_id) & (df['MAGNITUD'] == magnitud)]
```

```python
df['date'] = df['ANO'].astype(str) + '-' + df['MES'].astype(str) + '-' + df['DIA'].astype(str)
df['date'] = pd.to_datetime(df['date'])
```

```python
data = df.iloc[:,8:].copy()
data.set_index('date', inplace = True)
data = data.T.unstack()
```

```python
res = data.reset_index()
res = res[res.iloc[:,2]!='V'] # Drop validation rows
res['level_1'] = res['level_1'].map(lambda x: x[1:]).astype(int) # Remove "H" from hours prefix 
res['level_1'] =  pd.to_timedelta(res['level_1'], unit='h')
res['date'] = res['date']+res['level_1']
```

```python
res = res.rename(columns= {0:'Y'})
res.drop(columns= 'level_1', inplace = True)
```

```python
res['id'] = int('280790' + str(station_id))
```

```python
res = res.merge(stations[['lon','lat','elevation', 'id']], on='id')
res.set_index('date', inplace=True)
```

```python
res[res['Y']!= 'N']['Y'].plot()
```

```python
res = res[res['Y']!= 'N']
```

```python
res.to_csv('../data/result/sensor_result.csv')
```
