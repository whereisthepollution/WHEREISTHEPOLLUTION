```python
df_meteo = pd.read_csv('../data/result/meteo_result.csv')
df_sensor = pd.read_csv('../data/result/sensor_result.csv')
df_traffic = pd.read_csv('../data/result/traffic_result.csv')
```

## Add prefix to columns before merging

```python
df_meteo.columns = ['M_'+col if col != 'date' else col for col in df_meteo.columns.values]
df_traffic.columns = ['T_'+col if col != 'date' else col for col in df_traffic.columns.values]
df_sensor.columns = ['S_'+col if (col != 'date') and (col != 'Y') else col for col in df_sensor.columns.values]
```

```python
df_sensor.merge(df_traffic, on='date')
```

```python
df = df_sensor.merge(df_traffic, on='date').merge(df_meteo, on='date')
```

```python
df.set_index('date', inplace = True)
df.sort_index(inplace = True)
df.to_csv('../data/result/joined_result.csv')
```
